# Formatting with the.format() method

A good way to format objects into your strings for print statements is with the string. forma() method.


'String here {}then also{}'.format('something1', 'something2')

```python
print ('this is a string {}' .format('INSERTED'))
```

```python
print ('THE {} {} {}'.format('mimi','is','naughty'))
```

```python
print('The {m},{i},{n}'.format(m ='mimi',i= 'is', n='naughty'))
```
